# Python lists
What is a list?

An ordered collection of items, zero‑indexed (0, 1, 2, …).

Mutable: you can add/remove/modify items in place.

Often hold homogeneous items (e.g., a sequence of returns).


```python

# Create & index
prices = [100.0, 101.5, 100.8, 102.2]
first = prices[0]                 # 100.0
last_two = prices[-2:]            # [100.8, 102.2]

# Mutate
prices.append(103.0)              # [100.0, 101.5, 100.8, 102.2, 103.0]
prices[1] = 101.8                 # change in place

# Common utilities
n = len(prices)                   # count
delta = prices[-1] - prices[0]    # range

# List comprehension (compact transform)
returns = [(prices[i]/prices[i-1] - 1) for i in range(1, len(prices))]
```

# Python Dictionaries (key → value mappings)

What is a dict?

A mapping from keys to values (like a labelled lookup table).   
Keys must be unique and hashable (strings are perfect).   
Preserves insertion order (Python 3.7+).   
Ideal when labels matter more than positions.   

```python

# Define & access
weights = {"AUS_EQ": 0.40, "INTL_EQ": 0.30, "BONDS": 0.20, "RE": 0.05, "PEVC": 0.05}
aus_w = weights["AUS_EQ"]         # 0.40

# Update & safe access
weights["INTL_EQ"] = 0.28
maybe_cash = weights.get("CASH", 0.0)   # returns 0.0 if "CASH" not present
```

# Loops

## Why loops exist  
Repeat an action over a sequence (list, string, rows).  
Control flow with break / continue.  
Useful for I/O, conditional side‑effects, and step‑by‑step logic.  

## Core idea
Pure‑Python loops are flexible but often slower and more error‑prone for data transforms.
Prefer declarative operations that state what you want, not how to loop.

## When a loop is still the right tool
I/O & side effects (read files one-by-one, print/logging).
Stateful algorithms (carry‑forward logic, custom rule engines).
Clarity first: if a simple loop reads better than a contorted one‑liner, use the loop.

## Mental model 
Start with the highest‑level tool that fits (pandas/NumPy → comprehension → loop).
Loops are a fallback, not a first choice, for data transforms.



```python

# for: iterate items directly (preferred over range(len(...)))
prices = [100.0, 101.5, 100.8]
for p in prices:
    print(p)

# enumerate: need position + value
for i, p in enumerate(prices):
    print(i, p)

# range: numbers 0..n-1
for i in range(3):
    print(i)

# while: loop until a condition fails
count = 0
while count < 3:
    print("tick", count)
    count += 1
```

# Python Functions (small, reusable steps)

A named block of code you can reuse.  
It can take inputs (parameters) and return a value.  
Keeps notebooks tidy, reduces copy‑paste, and helps with testing later.  

```python

def greet(name):
    """Return a friendly greeting string."""
    return f"Welcome, {name}!"



def monthly_rate(annual_rate=0.04):
    """Convert annual rate to monthly (fractional)."""
    return annual_rate / 12


def pct_change(new, old):
    """Fractional change: (new - old) / old."""
    return (new - old) / old
```

```python
#mini self check
greet("FINM3422")
monthly_rate()          # uses default 0.04 → 0.003333...
pct_change(101.5, 100)  # 0.0
```

# Objects & OOP in Python

## What is an object?
A value with type (class), state (data), behavior (methods).  
In Python, everything is an object: numbers, strings, lists, dicts, functions, classes, modules.  
## Why OOP?
Encapsulation: keep related data + functions together.  
Reuse/extension: inheritance & composition.  
Polymorphism (duck typing): “works like a sequence? then len(x) works.”   

## Names vs objects
name → object (binding). The variable is the name; the object lives in memory.  
Rebinding a name doesn’t change other names pointing to the same object.  

## Essential object syntax (today):
Dot access: obj.attr  Method call: obj.method(...)  
Inspect: type(obj), dir(obj), help(obj)  
Mutability: lists/dicts mutable; numbers/strings immutable.  




```python

# Everything is an object
x = 3
type(x), id(x)           # (<class 'int'>, ...)

prices = [100, 101.5]
prices.append(102.2)     # method on a list object

def greet(n):            # function is an object too
    return f"Hi, {n}"
greet.__name__           # 'greet
```

```python
# mutability 
a = [1, 2]; b = a    # two names → the same list object
b.append(3)          # a now shows [1, 2, 3] as well
```

1- Dot notation to access attributes and call methods.  
2- type/dir/help to explore an unfamiliar object.  
3- Names bind to objects (not the other way around).  
4- Mutability matters (lists/dicts mutate; numbers/strings don’t).  
5- Functions are objects you can pass around
